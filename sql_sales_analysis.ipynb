```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("Sample - Superstore.csv")
df.head()
```

```python
conn = sqlite3.connect("sales.db")
df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT * FROM sales LIMIT 5;", conn)

```

```python
query = """
SELECT Category,
       ROUND(SUM(Sales), 2) AS total_sales
FROM sales
GROUP BY Category
ORDER BY total_sales DESC;
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT State,
       ROUND(SUM(Sales), 2) AS total_sales
FROM sales
GROUP BY State
ORDER BY total_sales DESC
LIMIT 5;
"""
top_states = pd.read_sql(query, conn)
top_states
```

```python
query = """
SELECT "Sub-Category",
       ROUND(AVG(Sales), 2) AS avg_sales
FROM sales
GROUP BY "Sub-Category"
ORDER BY avg_sales DESC;
"""
avg_sales = pd.read_sql(query, conn)
avg_sales.head(10)



```

```python
query = """
SELECT "Sub-Category",
       ROUND(SUM(Sales), 2) AS total_sales
FROM sales
GROUP BY "Sub-Category"
ORDER BY total_sales DESC
LIMIT 10;
"""
top_subcats = pd.read_sql(query, conn)
top_subcats
```

```python
## Key Insights

- Technology is the highest-revenue category by total sales.
- Sales are concentrated in a small number of states, led by California and New York.
- Sub-category performance differs by metric: Copiers rank highest by average sales per order, while Phones generate the highest total revenue.
```
